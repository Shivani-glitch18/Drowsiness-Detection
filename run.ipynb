```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from pygame import mixer
```

```python
# Load pre-trained model
model = load_model(r'C:\Users\rajat\OneDrive\Desktop\drowsiness\Project\trainmodels.keras')
```

```python
# Initialize face and eye cascade classifiers
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
# Initialize alarm sound
mixer.init()
sound = mixer.Sound(r'C:\Users\rajat\OneDrive\Desktop\drowsiness\alarm.wav')
```

```python
# Function to process each frame
def process_frame(frame):
    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    
    # Process each face
    for (x, y, w, h) in faces:
        # Extract eye region
        eye_region = gray[y:y+h, x:x+w]
        
        # Resize eye region and preprocess for model input
        eye = cv2.resize(eye_region, (224, 224))  # Resize to model input size
        eye = cv2.cvtColor(eye, cv2.COLOR_GRAY2RGB)  # Convert to RGB
        eye = np.expand_dims(eye, axis=0)  # Add batch dimension
        eye = eye / 255.0  # Normalize pixel values
        
        # Make prediction
        prediction = model.predict(eye)
        
        # Process prediction
        if prediction[0][0] > 0.30:  # Eyes closed
            # Perform action for eyes closed
            pass
        elif prediction[0][1] > 0.90:  # Eyes open
            # Perform action for eyes open
            pass
    
    return frame
```

```python
# Open webcam
cap = cv2.VideoCapture(0)

while True:
    # Read frame from webcam
    ret, frame = cap.read()
    
    # Check if frame was successfully captured
    if not ret:
        print("Error: Failed to capture frame from webcam")
        break
    
    # Process frame
    processed_frame = process_frame(frame)
    
    # Display frame
    cv2.imshow('frame', processed_frame)
    
    # Exit loop if 'q' is pressed
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break

# Release webcam and close windows
cap.release()
cv2.destroyAllWindows()
```
